```python
students_performance.groupby("lunch")[["math score", "reading score", "writing score"]]
```

```python
standart_lunch_query = students_performance["lunch"] == "standard"
free_red_lunch_query = students_performance["lunch"] == "free/reduced"
```

```python
lessons = list(students_performance.columns)[-3:]
for lesson in lessons:
    print(f"{lesson}")
    standart_lunch_mean = students_performance.loc[standart_lunch_query, lesson].mean()
    reduced_lunch_mean = students_performance.loc[free_red_lunch_query, lesson].mean()
    standart_lunch_var = students_performance.loc[standart_lunch_query, lesson].var()
    reduced_lunch_var = students_performance.loc[free_red_lunch_query, lesson].var()
    print(f"standard: {standart_lunch_mean};\t" f"reduced: {reduced_lunch_mean}")
    print(f"standard: {standart_lunch_var};\t" f"reduced: {reduced_lunch_var}")
```

```python
students_performance = students_performance.rename(
    columns={
        "parental level of education": "parental_level_of_education",
        "test preparation course": "test_preparation_course",
        "math score": "math_score",
        "reading score": "reading_score",
        "writing score": "writing_score",
    }
)
```
